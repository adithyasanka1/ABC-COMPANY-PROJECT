```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
myexcel=pd.read_csv("myexcel-ABC company.csv")
myexcel
```

## Replace incorrect data in the height coloumn with random numbers between 150 and 180

```python
myexcel['Height'] = np.random.uniform(150,180,size=len(myexcel))
print("Modified 'Height' Column")
print(myexcel['Height'])
```

## 1. How many are there in each team and the percentage splitting with respect to the Total Employees

```python
team_count=myexcel['Team'].value_counts()
team_percentage=(team_count/len(myexcel))*100
team_summary=pd.DataFrame({'count':team_count,'percentage':team_percentage})
team_summary
```

## 2.Segregate the employees with respect to different posotions

```python
position_count=myexcel['Position'].value_counts()
position_count
```

## 3. Find from which age group most of the employees belong to

```python
age=myexcel['Age'].value_counts
most_common_age=myexcel['Age'].mode()[0]
print("While comparing the age of employees,Most common age group is : ",most_common_age)
```

## 4.Findout from which team and position, spending in terms of salary is high

```python
high_salary_team_position=myexcel.groupby(['Team','Position'])['Salary'].mean()
spends_most=high_salary_team_position.idxmax()
spends_most
```

## Find if there is any corelation between age and salary,represent it visually

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Age',y='Salary',data=myexcel)
plt.title('Correlation between Age and Salary')
plt.show()
```

```python
plt.bar(list(myexcel['Age'].value_counts().keys()),list(myexcel['Age'].value_counts()),color=['blue','red'])
plt.title("Correlation between Age and Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
